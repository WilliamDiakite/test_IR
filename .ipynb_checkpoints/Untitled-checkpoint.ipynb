## Test co-occurrence graph (skill-job)


```python
import random
import csv

from pprint import pprint
from collections import defaultdict


# Create some jobs and soft skills

jobs1 = ['ingenieur', 'cuisinier', 'politicien']
jobs2 = ['mecanicien', 'boulanger', 'luthier']

skills1 = ['autonome', 'creatif', 'audacieux', 'curieux', 'sens du collectif']
skills2 = ['habile', 'motive', 'rigoureux', 'temeraire', 'ponctuel']


# Create some data
data = []
for i in range(10000):
    
    for j1, j2 in zip(jobs1, jobs2):
        
        c1 = (j1, random.choice(skills1))
        c2 = (j2, random.choice(skills2))
        data.append(c1)
        data.append(c2)
        
        if i % 7 == 0:
            c3 = (j1, random.choice(skills2))
            c4 = (j2, random.choice(skills1))
            data.append(c1)
            data.append(c2)
            
    
pprint(data)
```

```python
import operator

# Maintenant on calcul les co-occurrences pour chaque couple existant 
co_occ = defaultdict(lambda: 0)
for c in data:
    co_occ[c] += 1 
    
max_weight = sorted(co_occ.items(), key=operator.itemgetter(1))[-1][-1]
    
for k in co_occ:
    print('couple:', k, 'poids:', co_occ[k])
    
print('max:', max_weight)
    
```

```python
# On enregistre les donnes au format csv
with open('co_occurrence.csv', 'w') as f:
    header = ['Source', 'Target', 'Weight']
    writer = csv.DictWriter(f, fieldnames=header)
    writer.writeheader()
    
    for k in co_occ:
        row = {'Source': k[0], 'Target': k[1], 'Weight': co_occ[k]}
        writer.writerow(row)
```

```python
# On met les donnes sous forme de rseau grace  networkx
import networkx as nx
import matplotlib.pyplot as plt

g = nx.Graph()

for k in co_occ:
    job, skill = k[0], k[1]
    g.add_node(job, count=1)
    g.add_node(skill, count=1)
    g.add_edge(job, skill, weight=co_occ[(job, skill)])
nx.draw(g, with_labels = True)
plt.show()
```
